```python
# Fill all null values with "" and clean all non-perfect integers in 'affected' column
df = df.fillna("")
```

```python
#Find out how many complaints were filed by date
df['Created Date'].value_counts()
```

```python
#Strip the time element from 'Created Date' column strings
complaint_date = df["Created Date"].map(lambda x: str(x)[:-12])
complaint_date
```

```python
#Sort our values
complaint_date.value_counts().sort_values(ascending = False).reset_index()
```

```python
#Prepare our data for csv by renaming our indexes
date = complaint_date.value_counts().reset_index().rename(columns = {'index': 'Date', 'Created Date': 'Total Complaints'})
date = date[~date['Date'].str.contains('2019')]
date
```
